## Formulation of the problem

The following data are available for 1260 respondents:

* wage per hour of work, $;
* work experience, years;
* education, years;
* attractiveness, in points from 1 to 5;
* binary signs: gender, marital status, state of health (good / bad), labor union membership, skin color (white / black), employment in the service sector (yes / no).

It is required to evaluate the influence of attractiveness on the level of earnings taking into account all other factors.

Hamermesh D.S., Biddle J.E. (1994) Beauty and the Labor Market, American Economic Review, 84, 1174â€“1194.

Data:

```python
raw = pd.read_csv("beauty.csv", sep=";", index_col=False) 
raw.head()
```

```python
from pandas import plotting
pd.plotting.scatter_matrix(raw[['wage', 'exper', 'educ', 'looks']], alpha=0.2, 
                                 figsize=(15, 15), diagonal='hist')
pylab.show()
```

```python
print(raw.union.value_counts())
print(raw.goodhlth.value_counts())
print(raw.black.value_counts())
print(raw.female.value_counts())
print(raw.married.value_counts())
print(raw.service.value_counts())
```

## Preprocessing

```python
data = raw
```

Let's look at the distribution of the target attribute - the level of wages:

```python
plt.figure(figsize(5,7))
plt.subplot(211)
data['wage'].plot.hist()
plt.xlabel('Wage', fontsize=14)

plt.subplot(212)
np.log(data['wage']).plot.hist()
plt.xlabel('Log wage', fontsize=14)
pylab.show()
```

One person in the sample gets USD 77.72 per hour, the rest - less than USD 45; We will remove this person's observation so that the regression does not migrate to it.

```python
data = data[data['wage'] < 77]
```

Let's look at the distribution of estimates of attractiveness:

```python
plt.figure(figsize(8,7))
data.groupby('looks')['looks'].agg(lambda x: len(x)).plot(kind='bar', width=0.9)
plt.xticks(rotation=0)
plt.xlabel('Looks', fontsize=14)
pylab.show()
```

In the groups look = 1 and look = 5 too few observations. We turn the attribute into a categorical variable and encode it using dummy variables:

```python
data["looks"]
data['belowavg'] = data['looks'].apply(lambda x : 1 if x < 3 else 0)
data['aboveavg'] = data['looks'].apply(lambda x : 1 if x > 3 else 0)
data.drop('looks', axis=1, inplace=True)
```

The data is now:

```python
data.head()
```

## Model

### Simple model

Let us construct a linear model in all its features.

```python
m1 = smf.ols('wage ~ exper + union + goodhlth + black + female + married +'\
                    'service + educ + belowavg + aboveavg', 
             data=data)
fitted = m1.fit()
print(fitted.summary())
```

Let's look at the distribution of residuals:

```python
plt.figure(figsize(5,7))
plt.subplot(211)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(212)
np.log(fitted.resid).plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

It is skewed, like the original feature. In such situations, it often helps to move from a regression of the original characteristic to a regression of its logarithm.

### Logarithm

```python
m2 = smf.ols('np.log(wage) ~ exper + union + goodhlth + black + female + married +'\
                            'service + educ + belowavg + aboveavg', data=data)
fitted = m2.fit()
print(fitted.summary())

plt.figure(figsize(5,7))
plt.subplot(211)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(212)
np.log(fitted.resid).plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

Now it's better. Let us now look at the dependence of residuals on continuous features:

```python
plt.figure(figsize(5,7))
plt.subplot(211)
scatter(data['educ'],fitted.resid)
plt.xlabel('Education', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.subplot(212)
scatter(data['exper'],fitted.resid)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
pylab.show()
```

The second graph shows the quadratic dependence of the residuals from work experience. Let's try to add to the attributes a square of work experience to take into account this effect.

### Squared feature

```python
m3 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female +'\
                            'married + service + educ + belowavg + aboveavg', data=data)
fitted = m3.fit()
print(fitted.summary())

plt.figure(figsize(12,7))
plt.subplot(121)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(122)
np.log(fitted.resid).plot.hist()
plt.xlabel('Residuals', fontsize=14)
plt.figure(figsize(12,5))
plt.subplot(131)
scatter(data['educ'],fitted.resid)
plt.xlabel('Education', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.subplot(132)
scatter(data['exper'],fitted.resid)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.subplot(133)
scatter(data['exper']**2,fitted.resid)
plt.xlabel('Experience^2', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
pylab.show()
```

We use the Breusch-Pagan test to to test for heteroskedasticity of error:

```python
print('Breusch-Pagan test: p = %f' % sms.het_breuschpagan(fitted.resid, fitted.model.exog)[1])
# If the test statistic has a p-value below an appropriate threshold (e.g. p<0.05) 
# then the null hypothesis of homoskedasticity is rejected and heteroskedasticity assumed.
```

Errors are heteroscedastic, therefore, the significance of the signs can be determined incorrectly. Let us make the White's correction (White, 1980):

```python
m4 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female +'\
                            'married + service + educ + belowavg + aboveavg', data=data)
fitted = m4.fit(cov_type='HC1')
print(fitted.summary())

plt.figure(figsize(5,7))
plt.subplot(211)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(212)
np.log(fitted.resid).plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

### Remove insignificant features

In the previous model the following are insignificant: skin color, health, marital status. Let's remove them. The indicator of attractiveness above the average is also insignificant, but we will not delete it, because this is one of the variables by which we need to answer the question at the end.

```python
m5 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + female + service + educ +'\
                            'belowavg + aboveavg', data=data)
fitted = m5.fit(cov_type='HC1')
print(fitted.summary())

plt.figure(figsize(5,7))
plt.subplot(211)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(212)
np.log(fitted.resid).plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

Let's see if the model after the removal of the three features is significantly worse, with the Fisher criterion:

```python
print("F=%f, p=%f, k1=%f" % m4.fit().compare_f_test(m5.fit()))
```

It is ok.

Let's check if there are any observations that influence the regression equation too much:

```python
plt.figure(figsize(8,7))
plot_leverage_resid2(fitted)
pylab.show()
```

## Results

The resulting model explains the 40% variation of the logarithm of the response.

```python
plt.figure(figsize(5,7))
plt.subplot(211)
scatter(data['wage'],np.exp(fitted.fittedvalues))
plt.xlabel('Wage', fontsize=14)
plt.ylabel('Exponentiated predictions', fontsize=14)
plt.xlim([0,50])

plt.subplot(212)
scatter(np.log(data['wage']),fitted.fittedvalues)
plt.xlabel('Log wage', fontsize=14)
plt.ylabel('Predictions', fontsize=14)
plt.xlim([0,4])
pylab.show()
```

For factors of interest to us, the coefficients are -0.1307 (below average) and -0.0010 (above average). 

Since the regression was made to the logarithm of the response, they can be interpreted as a percentage increase. Taking into account additional factors, representatives of the general population from which the sample is taken receive, on average:

* 13% less if their attractiveness is below the average (p = 0.001, 95% confidence interval - [5.21]%);
* the same if their attractiveness is above the average (p = 0.972, 95% confidence interval - [-6.6]%).
